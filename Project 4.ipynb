```python
import requests
from bs4 import BeautifulSoup
from time import sleep
r = requests.get("https://results.chicagomarathon.com/well-known/2019/?content=well-known&lang=EN_CAP&num_results=50&pid=list&search%5Bsex%5D=M&search%5Bage_class%5D=%25&event=MAR&favorite_add=999999107FA317000023A4AD")
```

```python
soup = BeautifulSoup(r.text)
soup
```

```python
s = soup.find_all("ul", class_="list-group list-group-multicolumn")
```

```python
len(s)
```

```python
# the first li is a header not a runner link.
runners = s[0].find_all("li")[1:]
runners
```

```python
## Link for a single runner.
runners[0].find('a')["href"]
```

```python
## requesting runners information from the website
string = "https://results.chicagomarathon.com/well-known/2019/"
r_list = []
for i in range(len(runners)):
    url = string+runners[i].find('a')["href"]
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    r_list.append(soup)
    sleep(2)
```

```python
r_list[0]
```

```python
## The last one is the short initial which is not needed.
r_list[0].find("table").find_all("tr")[:-1]
```

```python
# create a df, start with runners 'personal' information
import pandas as pd
df= pd.DataFrame()
df["Name (CTZ)"] = [runner.find("table").find_all("tr")[:-1][0].find("td").get_text() for runner in r_list]
df["Age Group"] = [runner.find("table").find_all("tr")[:-1][1].find("td").get_text() for runner in r_list]
df["Bib Number"] = [runner.find("table").find_all("tr")[:-1][2].find("td").get_text() for runner in r_list]
df["Age"] = [runner.find("table").find_all("tr")[:-1][3].find("td").get_text() for runner in r_list]
df["City, State"] = [runner.find("table").find_all("tr")[:-1][4].find("td").get_text() for runner in r_list]
```

```python
df
```

```python
## A list of all the time split for one runner
r_list[0].find("table", class_="table table-condensed table-striped").find_all("td", class_="time")
```

```python
## time split for all runners
df["05K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[0].get_text() for runner in r_list]
df["10K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[1].get_text() for runner in r_list]
df["15K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[2].get_text() for runner in r_list]
df["20K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[3].get_text() for runner in r_list]
df["HALF"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[4].get_text() for runner in r_list]
df["25K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[5].get_text() for runner in r_list]
df["30K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[6].get_text() for runner in r_list]
df["35K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[7].get_text() for runner in r_list]
df["40K"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[8].get_text() for runner in r_list]
df["FINISH"] = [runner.find("table", class_="table table-condensed table-striped").find_all("td", class_="time")[9].get_text() for runner in r_list]
```
